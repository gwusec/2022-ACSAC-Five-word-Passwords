```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
```

```python
color_2 = ['#abd9e9','#fdae61']
```

```python
post_recall = pd.DataFrame(data={'All Participants':['Treatment 1', 'Treatment 2', 'Treatment 3', 'Overall'], \
                                 'Recall':[47.37, 23.68, 47.50, 39.66], 'No Recall': [52.63, 76.32, 52.50, 60.34]\
                                }).set_index('All Participants')

post_recall = post_recall.iloc[::-1]
post_recall_num = [46, 19, 9, 18, 70, 21, 29, 20]

post_recall_no_help = pd.DataFrame(data={'No External Help':['Treatment 1', 'Treatment 2', 'Treatment 3', 'Overall'], \
                                         'Recall':[0, 7.14, 8.33, 5.56], 'No Recall': [100, 92.86, 91.67, 94.44]\
                                        }).set_index('No External Help')

post_recall_no_help = post_recall_no_help.iloc[::-1]
post_recall_no_help_num = [2, 1, 1, 0, 32, 10, 12, 10]

post_recall_help = pd.DataFrame(data={'With External Help':['Treatment 1', 'Treatment 2', 'Treatment 3', 'Overall'], \
                                         'Recall':[64.29, 32, 62.07, 53.65], 'No Recall': [35.71, 68, 37.93, 46.35]\
                                        }).set_index('With External Help')

post_recall_help = post_recall_help.iloc[::-1]
post_recall_help_num = [44, 18, 8, 18, 38, 11, 17, 10]
```

```python
import matplotlib

matplotlib.use("pdf")
matplotlib.rcParams.update({
    "pgf.texsystem": "pdflatex",
    'font.family': 'sans-serif',
    'font.sans-serif': 'Arial',
    'text.usetex': True,
    'pgf.rcfonts': False,
    'font.size': 14
})
```

```python
# Create a figure.
fig = plt.figure(figsize=(6.8, 7.2))

# Divide the figure into a 1x4 grid, and give me the first section
ax1 = fig.add_subplot(3, 1, 1)
ax2 = fig.add_subplot(312, sharex=ax1)
ax3 = fig.add_subplot(313, sharex=ax1)

plt.rcParams['figure.dpi'] = 800
plt.rcParams['savefig.dpi'] = 800

post_recall.plot(kind='barh', stacked=True, color = color_2, ax=ax1, width=0.8, zorder=3)
# ax1.xaxis.set_major_formatter(mtick.PercentFormatter())
# ax1.tick_params(axis='x', labelrotation = 0, size=14)
# ax1.get_xaxis().set_visible(False)
ax1.tick_params(axis='x', colors='silver')
ax1.tick_params(axis='y', labelrotation = 0, size=16)

ax1.spines['top'].set_visible(False)
ax1.spines['right'].set_visible(False)
ax1.spines['bottom'].set_visible(False)

index = 0

for bar in ax1.patches:
    ax1.text(
        bar.get_x() + bar.get_width() / 2,
        bar.get_y()+ bar.get_height() / 2,
        post_recall_num[index],
        ha='center',
        va='center',
        size=14
    )
    
    index = index + 1
    
ax1.legend(loc='upper center', prop={'size':14}, ncol=2, bbox_to_anchor=(0.5, 1.3), frameon=False)

ax1.yaxis.label.set_fontsize(16)
ax1.grid(axis='x', zorder=0)

post_recall_no_help.plot(kind='barh', stacked=True, color = color_2, ax=ax2, width=0.8, zorder=3)
ax2.xaxis.set_major_formatter(mtick.PercentFormatter())

ax2.tick_params(axis='x', colors='silver')
ax2.tick_params(axis='y', labelrotation = 0, size=16)

ax2.spines['top'].set_visible(False)
ax2.spines['right'].set_visible(False)
ax2.spines['bottom'].set_visible(False)

index = 0

for bar in ax2.patches:
    if(post_recall_no_help_num[index] != 0):
        ax2.text(
            bar.get_x() + bar.get_width() / 2,
            bar.get_y()+ bar.get_height() / 2,
            post_recall_no_help_num[index],
            ha='center',
            va='center',
            size=14
        )
    
    index = index + 1
    
ax2.legend().set_visible(False)

ax2.yaxis.label.set_fontsize(16)
ax2.grid(axis='x', zorder=0)

post_recall_help.plot(kind='barh', stacked=True, color = color_2, ax=ax3, width=0.8, zorder=3)
ax3.xaxis.set_major_formatter(mtick.PercentFormatter())
# ax2.tick_params(axis='x', labelrotation = 0, size=14)
# ax2.get_xaxis().set_visible(False)
ax3.tick_params(axis='y', labelrotation = 0, size=16)

ax3.spines['top'].set_visible(False)
ax3.spines['right'].set_visible(False)

index = 0

for bar in ax3.patches:
    if(post_recall_help_num[index] != 0):
        ax3.text(
            bar.get_x() + bar.get_width() / 2,
            bar.get_y()+ bar.get_height() / 2,
            post_recall_help_num[index],
            ha='center',
            va='center',
            size=14
        )
    
    index = index + 1
    
ax3.legend().set_visible(False)

ax3.yaxis.label.set_fontsize(16)
ax3.grid(axis='x', zorder=0)

plt.subplots_adjust(wspace=0, hspace=0.1)

plt.savefig('post_recall_all.pdf', bbox_inches='tight', dpi=800)
```
