```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import matplotlib

matplotlib.use("pdf")
matplotlib.rcParams.update({
    "pgf.texsystem": "pdflatex",
    'font.family': 'sans-serif',
    'font.sans-serif': 'Arial',
    'text.usetex': True,
    'pgf.rcfonts': False,
    'font.size': 14
})
```

```python
word_freq_all = pd.DataFrame(data={'All Treatments':['this','love','the','could','cat',\
                                            'blue','water','and','can','very',\
                                            'dog','girl','six','your','letter',\
                                            'pair','dance','wet','strip','often',\
                                            'desk','you','one','fish','will',\
                                            'way','walk','third','three','play'],\
                                   'Cnt':[16,9,6,6,5,5,5,5,4,4,\
                                          4,4,4,4,3,3,3,3,3,3,\
                                          3,3,3,3,3,3,3,3,3,3]}).set_index('All Treatments')
word_freq_all = word_freq_all.iloc[::-1]

word_freq_tre1 = pd.DataFrame(data={'Treatment 1':['escape','letter','pair','ban','common',\
                                             'core','cow','cycle','dad','dance',\
                                             'evolve','girl','heat','six','third',\
                                             'tube','walk','waste','way','wet'],\
                                    'Cnt': [3,3,3,2,2,2,2,2,2,2,\
                                            2,2,2,2,2,2,2,2,2,2]}).set_index('Treatment 1')

# word_freq_tre1 = word_freq_tre1.iloc[::-1]

word_freq_tre2 = pd.DataFrame(data={'Treatment 2':['mood','belief','broken','like','often',\
                                             'singer','strip','stroke','teach'],\
                                    'Cnt': [3,2,2,2,2,2,2,2,2]}).set_index('Treatment 2')
# word_freq_tre2 = word_freq_tre2.iloc[::-1]

word_freq_tre3 = pd.DataFrame(data={'Treatment 3':['this','love','could','the','and',\
                                                   'water','cat','your','blue','can',\
                                                   'dog','too','very','will','you',\
                                                   'always','animal','apple','become','bird',\
                                                   'code', 'cool', 'eat', 'fish', 'five',\
                                                   'fly', 'food', 'four', 'great', 'green',\
                                                   'happy', 'money', 'much', 'orange', 'really', \
                                                   'safe', 'six', 'sky', 'smart', 'super', \
                                                   'sweet', 'today', 'world'],\
                                    'Cnt': [15,9,6,6,5,5,4,4,3,3,\
                                            3,3,3,3,3,2,2,2,2,2,\
                                            2,2,2,2,2,2,2,2,2,2,\
                                            2,2,2,2,2,2,2,2,2,2,\
                                            2,2,2]}).set_index('Treatment 3')
# word_freq_tre3 = word_freq_tre3.iloc[::-1]
```

```python
# Create a figure.
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 4), gridspec_kw={'width_ratios': [20, 9.3]})

# Divide the figure into a 1x4 grid, and give me the first section
# ax1 = fig.add_subplot(121)
# ax2 = fig.add_subplot(122)

# ax2.tick_params(axis='y', labelleft=False, labelright=True)

plt.rcParams['figure.dpi'] = 800
plt.rcParams['savefig.dpi'] = 800

from matplotlib.ticker import MaxNLocator

# ax1 = plt.figure().gca()
ax1.yaxis.set_major_locator(MaxNLocator(integer=True))
ax2.yaxis.set_major_locator(MaxNLocator(integer=True))
ax2.yaxis.tick_right()


# word_freq_all.plot(kind='barh', color = '#d7191c', ax=ax1, zorder=3)
word_freq_tre1.plot(kind='bar', color = '#fdae61', ax=ax1, zorder=3, width=0.8)
word_freq_tre2.plot(kind='bar', color = '#abd9e9', ax=ax2, zorder=3, width=0.8)

# new_list = range(0, 4)
# plt.xticks(new_list)

# ax1.tick_params(axis='x', labelrotation = 45)
# ax2.tick_params(axis='x', labelrotation = 45)

ax1.set_xticklabels(word_freq_tre1.index.values.tolist(), rotation=45, ha='right')
ax2.set_xticklabels(word_freq_tre2.index.values.tolist(), rotation=45, ha='right')

ax1.spines['top'].set_visible(False)
ax1.spines['right'].set_visible(False)
ax2.spines['top'].set_visible(False)
# ax2.spines['right'].set_visible(False)
ax2.spines['left'].set_visible(False)

ax1.legend().set_visible(False)
ax2.legend().set_visible(False)
    
# ax.legend(loc='upper center', prop={'size':12}, ncol=2, bbox_to_anchor=(0.5, 1.1), frameon=False)

ax1.grid(axis='y', zorder=0)
ax2.grid(axis='y', zorder=0)

plt.subplots_adjust(wspace=0.025, hspace=0)
plt.minorticks_off()
# plt.show()
plt.savefig('word_freq_12.pdf', bbox_inches='tight')
```
