```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
```

```python
import matplotlib

matplotlib.use("pdf")
matplotlib.rcParams.update({
    "pgf.texsystem": "pdflatex",
    'font.family': 'sans-serif',
    'font.sans-serif': 'Arial',
    'text.usetex': True,
    'pgf.rcfonts': False,
    'font.size': 12
})
```

```python
color_5 = ['#d7191c', '#fdae61','#ffffbf','#abd9e9','#2c7bb6']
```

```python
# Confident Secure
confident_secure = pd.DataFrame(data={'How Confident Secure':['Security', 'Memorability'], \
                                      'Not confident': [9.48, 34.48], \
                                      'Slightly confident': [11.20, 30.17], \
                                      'Moderately confident': [31.90, 18.97], \
                                      'Confident': [32.76, 10.34], \
                                      'Very confident': [14.66, 6.03]}).set_index('How Confident Secure')

confident_secure_num = [11, 40, 13, 35, 37, 22, 38, 12, 17, 7]

# # Confident Remember
# confident_remember = pd.DataFrame(data={'How Confident Remember':['Not confident', 'Slightly confident', \
#                                                                'Moderately confident', 'Confident', 'Very confident'], \
#                                      'Percentage':[34.48, 30.17, 18.97, 10.34, 6.03]}).set_index('How Confident Remember')

# confident_remember_num = [40, 35, 22, 12, 7]
```

```python
# Create a figure.
fig = plt.figure(figsize=(6.8, 1.7))

# Divide the figure into a 1x4 grid, and give me the first section
ax1 = fig.add_subplot(111)

# ax2.tick_params(axis='y', labelleft=False, labelright=True)

plt.rcParams['figure.dpi'] = 800
plt.rcParams['savefig.dpi'] = 800

from matplotlib.ticker import MaxNLocator

# ax1 = plt.figure().gca()
ax1.xaxis.set_major_formatter(mtick.PercentFormatter())


# word_freq_all.plot(kind='barh', color = '#d7191c', ax=ax1, zorder=3)
confident_secure.plot(kind='barh', stacked=True, color = color_5, \
                        ax=ax1, zorder=3, width=0.7)

ax1.set_yticklabels(confident_secure.index.values.tolist())

ax1.spines['top'].set_visible(False)
ax1.spines['right'].set_visible(False)

index = 0

for bar in ax1.patches:
    ax1.text(
        bar.get_x() + bar.get_width() / 2,
        bar.get_y()+ bar.get_height() / 2,
        confident_secure_num[index],
        ha='center',
        va='center',
        size=14
    )
    
    index = index + 1

ax1.grid(axis='x', zorder=0)
ax1.get_yaxis().get_label().set_visible(False)

# ax1.legend(loc='upper center', prop={'size':14}, ncol=3, bbox_to_anchor=(0.4, 1.55), frameon=False)

# reordering the labels
handles, labels = plt.gca().get_legend_handles_labels()
  
# specify order
order = [0, 3, 1, 4, 2]
  
# pass handle & labels lists along with order as below
ax1.legend([handles[i] for i in order], [labels[i] for i in order],\
           loc='upper center', prop={'size':14}, ncol=3, bbox_to_anchor=(0.4, 1.55), frameon=False)

plt.subplots_adjust(wspace=0.15, hspace=0)
plt.minorticks_off()
# plt.show()
plt.savefig('confident_secure_remember.pdf', bbox_inches='tight')
```
